```python
import pandas as pd
```

```python
merged_player_games = pd.read_csv(r'..\merg\merged_player_games.csv')
merged_player_games['appearance'] = merged_player_games['played_minutes'].notnull().astype(int)
merged_player_games['appearance'].unique()
```

```python
merged_player_games['appearance'].value_counts()
```

```python
merged_games = pd.read_csv(r'..\merg\merged_games.csv')
merged_games['date'] = pd.to_datetime(merged_games['date'])
print([d.year if (d.month>=8) else (d.year - 1) for d in merged_games['date']])
```

```python
merged_games['Year'] = [d.year if (d.month>=8) else (d.year - 1) for d in merged_games['date']]
merged_games
```

```python
merged_df = pd.merge(merged_games, merged_player_games, on='game_id')
merged_df = merged_df[(merged_df['Year'] >= 2015) & (merged_df['Year'] <= 2021)]
```

```python
# group the DataFrame by year and calculate the sum of the Value column for each year
grouped_df = merged_df.groupby(['player_id', 'Year'])['appearance'].sum().reset_index()

# rename the columns
grouped_df.columns = ['player_id', 'Year', 'appearance sum']

# print the grouped DataFrame
print(grouped_df)
```

```python
grouped_df.to_csv('grouped_df.csv', index=False)
```

```python
merged_player_ownGoals = pd.read_csv(r'..\merg\merged_player_ownGoals.csv')
```

```python
merged_df = pd.merge(merged_df, merged_player_ownGoals, on='game_id')
```

```python
# group the DataFrame by year and calculate the sum of the Value column for each year
grouped_df2 = merged_df.groupby(['player_id', 'Year'])['own_goals'].sum().reset_index()

# rename the columns
grouped_df2.columns = ['player_id', 'Year', 'own_goals']

# print the grouped DataFrame
print(grouped_df2)
```

```python
mergemerge = pd.concat([grouped_df, grouped_df2])
mergemerge
```

```python
mergemerge.to_csv('mergemerge.csv', index=False)
```
